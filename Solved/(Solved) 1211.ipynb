```python
queries["quality"] = queries["rating"] / queries["position"] + 1e-10
queries["poor_quality"] = queries["rating"].apply(lambda x: 1 if x < 3 else 0)
composite_quality = (
    queries.groupby("query_name")
    .agg({"quality": "sum", "poor_quality": "sum"})
    .reset_index()
)
count_num = queries.query_name.value_counts().reset_index()
composite = pd.merge(composite_quality, count_num, on="query_name", how="outer")
composite["poor_query_percentage"] = round(
    (composite["poor_quality"] / composite["count"]) * 100, 2
)
composite["quality"] = round(composite["quality"] / composite["count"], 2)
composite = composite.drop(columns=["poor_quality", "count"])
composite.sort_values(by="query_name", ascending=False)
```
